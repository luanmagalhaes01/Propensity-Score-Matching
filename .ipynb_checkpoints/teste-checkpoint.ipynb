```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo do gráfico
sns.set(style="whitegrid")

# Lê o arquivo CSV e armazena no dataframe 'ecls'
ecls = pd.read_csv("ecls.csv")

# Seleciona colunas específicas do dataframe 'ecls'
ecls = ecls[['catholic', 'c5r2mtsc_std', 'p5hmage', 
              'race_white', 'w3income_1k', 'p5numpla', 'w3momed_hsb']]

# Exibe as duas primeiras linhas do dataframe
print(ecls.head(2))
```

```python
# Calculando a média das notas de matemática
mean_math = ecls.groupby('catholic').agg(
    n_students=('c5r2mtsc_std', 'size'),
    mean_math=('c5r2mtsc_std', 'mean'),
    std_error=('c5r2mtsc_std', lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
).reset_index()

print(mean_math)
```

```python
# Calculando a média das variáveis de covariáveis
covariates = ['race_white', 'p5hmage', 'w3income_1k', 'p5numpla', 'w3momed_hsb']

mean_covariates = ecls.groupby('catholic')[covariates].mean().reset_index()
print(mean_covariates)
```

```python
# Modelagem do Propensity Score com regressão logística
X = ecls[['race_white', 'w3income_1k', 'p5hmage', 'p5numpla', 'w3momed_hsb']]
y = ecls['catholic']

logit_model = sm.Logit(y, sm.add_constant(X)).fit()
print(logit_model.summary())
```

```python
# Prevendo os propensity scores
ecls['pr_score'] = logit_model.predict(sm.add_constant(X))

# Visualizando a distribuição dos propensity scores
plt.figure(figsize=(15, 5))
sns.histplot(data=ecls, x='pr_score', hue='catholic', element='step', stat='density', common_norm=False, bins=30)
plt.xlabel('Probabilidade de estudar em escola católica')
plt.ylabel('Densidade')
plt.title('Distribuição dos Propensity Scores')
plt.legend(title='Escola em que cursou ensino médio', labels=['Católica', 'Não católica'])
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

# Separar grupos
treatment_group = ecls[ecls['catholic'] == 1]
control_group = ecls[ecls['catholic'] == 0]

# Selecionar variáveis independentes para matching
variables = ['race_white', 'w3income_1k', 'p5hmage', 'p5numpla', 'w3momed_hsb']
#variables = ['pr_score']

# Treinar o modelo de Nearest Neighbors
nn = NearestNeighbors(n_neighbors=1)
nn.fit(control_group[variables])

# Encontrar os vizinhos mais próximos para o grupo de tratamento
distances, indices = nn.kneighbors(treatment_group[variables])

# Criar um DataFrame de matched data
matched_control = control_group.iloc[indices.flatten()]

# Criar a nova base de dados combinando tratamento e controle
matched_data = pd.concat([treatment_group.reset_index(drop=True), matched_control.reset_index(drop=True)], axis=0)

print(matched_data)
```

```python
# Calculando a média das variáveis de covariáveis
covariates = ['race_white', 'p5hmage', 'w3income_1k', 'p5numpla', 'w3momed_hsb','c5r2mtsc_std']

mean_covariates_matched = matched_data.groupby('catholic')[covariates].mean().reset_index()
print(mean_covariates_matched)
```

```python
from scipy import stats


# Teste t após o matching
t_test_after_matching = stats.ttest_ind(
    matched_data[matched_data['catholic'] == 1]['c5r2mtsc_std'],
    matched_data[matched_data['catholic'] == 0]['c5r2mtsc_std'],
    equal_var=False
)

print(f"T-statistic after matching: {t_test_after_matching.statistic}, P-value: {t_test_after_matching.pvalue}")
```
